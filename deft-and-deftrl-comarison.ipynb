```python
import matplotlib.pyplot as plt
import numpy as np

# Set the data
trials = np.arange(1, 7)
deft_latency = [18.54264, 17.91638, 17.0489, 17.48368, 15.82598, 15.44298]
deftrl_latency = [8.73084, 13.48116, 9.3287, 12.06178, 7.8545, 12.68132]

# Create the figure and axis with a larger size for publication
plt.figure(figsize=(10, 6), dpi=300)

# Create the line plot with enhanced styling
plt.plot(trials, deft_latency, marker='o', linestyle='-', linewidth=2, 
         color='#1E90FF', label='DEFT', markersize=8, markerfacecolor='white', 
         markeredgewidth=2, markeredgecolor='#1E90FF')
plt.plot(trials, deftrl_latency, marker='s', linestyle='-', linewidth=2, 
         color='#FF6347', label='DeftRL', markersize=8, markerfacecolor='white', 
         markeredgewidth=2, markeredgecolor='#FF6347')

# Customize the plot for a professional look
plt.title('Latency Comparison: DEFT vs DeftRL', fontsize=16, fontweight='bold')
plt.xlabel('Trials', fontsize=12)
plt.ylabel('Latency (ms)', fontsize=12)
plt.grid(True, linestyle='--', linewidth=0.5, color='grey', alpha=0.7)

# Add a legend with a frame
plt.legend(frameon=True, framealpha=0.8, edgecolor='lightgrey')

# Annotate mean values
deft_mean = np.mean(deft_latency)
deftrl_mean = np.mean(deftrl_latency)
plt.axhline(y=deft_mean, color='#1E90FF', linestyle='--', alpha=0.5)
plt.axhline(y=deftrl_mean, color='#FF6347', linestyle='--', alpha=0.5)

plt.text(6.3, deft_mean, f'DEFT Mean: {deft_mean:.2f} ms', 
         verticalalignment='center', color='#1E90FF')
plt.text(6.3, deftrl_mean, f'DeftRL Mean: {deftrl_mean:.2f} ms', 
         verticalalignment='center', color='#FF6347')

# Tight layout to ensure all elements are visible
plt.tight_layout()

# Save the figure with high resolution
plt.savefig('deft_vs_deftrl_latency.png', dpi=300, bbox_inches='tight')

# Show the plot
plt.show()
```
